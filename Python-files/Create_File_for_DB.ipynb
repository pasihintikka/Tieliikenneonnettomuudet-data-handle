# Create File for SQL database

Search the information about population of the provinces (Maakunnat) by years from Tilastokeskus open data. After that merge information into Maakunnat data used in Onnettomuudet db's Alueet table. So can create new table Väestö and store new data there. 

```python
# Used libraries
import pandas as pd
import numpy as np
```

```python
# File path and name
src_file = '../datasets/onnettomuus/db/väestö_maakunnittain.csv'
target_file = '../datasets/onnettomuus/db/väestö_maakunnittain.csv'
result_file = '../datasets/onnettomuus/db/väestö_maakunnittain_db.csv'

# File attributes
src_delimiter = ';'
src_encoding = 'utf-8'
rslt_delimiter = ';'
rslt_encoding = 'utf-8'
```

### Download source files

```python
# Information about population of the Maakunnat by years
df = pd.read_csv(src_file, delimiter=src_delimiter, encoding=src_encoding)
```

```python
df
```

```python
# Used Maakunnat data in Alueet table
df_m = pd.read_csv('../datasets/onnettomuus/db/maakunnat.csv', delimiter=src_delimiter, encoding=src_encoding)
df_m
```

```python
# Change index for looping data
df.set_index(df.columns[0], inplace=True)
df
```

```python
# Looping data values into lists for creating new dataframe
value1 = [] # Väestö amount from table
value2 = [] # Maakuntsel name from table header
value3 = [] # Vuosi value from table index

for index in range(df.shape[1]): # Loop through columns
    for i in range(len(df)):     # Loop through rows
        value1.append(df.iloc[i][index])
        value2.append(df.iloc[i:i+1,index].name)
        value3.append(df.index[i])
```

```python
# Insert data values lists into dataframe columns
df_t = pd.DataFrame(columns=['Väestö','Maakuntsel','Vuosi'])
df_t['Väestö'] = pd.DataFrame(value1)
df_t['Maakuntsel'] = pd.DataFrame(value2)
df_t['Vuosi'] = pd.DataFrame(value3)
```

```python
df_t
```

```python
# Merge dataframes into result using column with the same content
df_result = pd.merge(df_m,df_t,on='Maakuntsel')
```

```python
df_result
```

```python
# Create result csv file for inserting database
df_result.to_csv(result_file, sep=rslt_delimiter, encoding=rslt_encoding, index=False)
```

# Create File for SQL database

Search the information about population of the provinces (Maakunnat) by years from Tilastokeskus open data. After that merge information into Maakunnat data used in Onnettomuudet db's Alueet table. So can create new table Väestö and store new data there. 

```python
# Used libraries
import pandas as pd
import numpy as np
```

```python
# File path and name
src_file = '../datasets/onnettomuus/db/väestö_maakunnittain.csv'
target_file = '../datasets/onnettomuus/db/väestö_maakunnittain.csv'
result_file = '../datasets/onnettomuus/db/väestö_maakunnittain_db.csv'

# File attributes
src_delimiter = ';'
src_encoding = 'utf-8'
rslt_delimiter = ';'
rslt_encoding = 'utf-8'
```

### Download source files

```python
# Information about population of the Maakunnat by years
df = pd.read_csv(src_file, delimiter=src_delimiter, encoding=src_encoding)
```

```python
df
```

```python
# Used Maakunnat data in Alueet table
df_m = pd.read_csv('../datasets/onnettomuus/db/maakunnat.csv', delimiter=src_delimiter, encoding=src_encoding)
df_m
```

```python
# Change index for looping data
df.set_index(df.columns[0], inplace=True)
df
```

```python
# Looping data values into lists for creating new dataframe
value1 = [] # Väestö amount from table
value2 = [] # Maakuntsel name from table header
value3 = [] # Vuosi value from table index

for index in range(df.shape[1]): # Loop through columns
    for i in range(len(df)):     # Loop through rows
        value1.append(df.iloc[i][index])
        value2.append(df.iloc[i:i+1,index].name)
        value3.append(df.index[i])
```

```python
# Insert data values lists into dataframe columns
df_t = pd.DataFrame(columns=['Väestö','Maakuntsel','Vuosi'])
df_t['Väestö'] = pd.DataFrame(value1)
df_t['Maakuntsel'] = pd.DataFrame(value2)
df_t['Vuosi'] = pd.DataFrame(value3)
```

```python
df_t
```

```python
# Merge dataframes into result using column with the same content
df_result = pd.merge(df_m,df_t,on='Maakuntsel')
```

```python
df_result
```

```python
# Create result csv file for inserting database
df_result.to_csv(result_file, sep=rslt_delimiter, encoding=rslt_encoding, index=False)
```

# Insert corrected coordinates into files for SQL database

#### Define parameters of the source and result files

```python
# Used libraries
import pandas as pd
import numpy as np
import math
```

```python
# File path and name
year = 2005
src_file = '../datasets/onnettomuus/co/tielonnett_'+str(year)+'_onnett.csv'
db_file = '../datasets/onnettomuus/db/tielonnett_'+str(year)+'_onnett.csv'
result_file = '../datasets/onnettomuus/db/tielonnett_'+str(year)+'_onnett.csv'
# File attributes
src_delimiter = ';'
src_encoding = 'utf-8'
rslt_delimiter = ';'
rslt_encoding = 'utf-8'
```

```python
df_src = pd.read_csv(src_file, delimiter=src_delimiter, encoding=src_encoding)
```

```python
df_db = pd.read_csv(db_file, delimiter=src_delimiter, encoding=src_encoding)
```

```python
df_co_src = pd.DataFrame(df_src, columns = ['Onnett_id','position.lat','position.lon'])
#df_co_src
```

```python
df_db_src = pd.DataFrame(df_db, columns = ['Onnett_id','position.lat','position.lon'])
#df_db_src
```

#### Collect source file rows which contains coordination values

```python
coord = []
coord = df_co_src[(df_co_src['position.lat'] > 0)]
df_co_lalo = pd.DataFrame(coord)
```

```python
df_co_lalo
```

```python
coord_db = []
coord_db = df_db_src[(df_db_src['position.lat'] > 0)]
df_db_lalo = pd.DataFrame(coord_db)
```

```python
df_db_lalo
```

#### Replace old values to db dataframe with corrected values by dataframe index

```python
#get old and new values using new frame row id
for i in df_co_lalo.index:
    # replace values of one DataFrame with the value of another DataFrame
    df_db['position.lat'] = df_db['position.lat'].replace(df_db['position.lat'][i],df_co_lalo['position.lat'][i])
    #print('row: ', i, ' X:', ' old: ',df_db['position.lat'][i],' new: ',df_co_lalo['position.lat'][i])
    df_db['position.lon'] = df_db['position.lon'].replace(df_db['position.lon'][i],df_co_lalo['position.lon'][i])
    #print('row: ', i, ' Y:', ' old: ',df_db['position.lon'][i],' new: ',df_co_lalo['position.lon'][i])
```

```python
for i in df_co_lalo.index:
    print('row: ', i, ' lat:', df_db['position.lat'][i],' lon: ',df_db['position.lon'][i])
```

#### Checking unchanged values

```python
coord_er = []
coord_er = df_db[(df_db['position.lat'] < 50)]
df_db_err = pd.DataFrame(coord_er)
```

```python
df_db_err = pd.DataFrame(df_db_err, columns = ['Onnett_id','position.lat','position.lon'])
df_db_err
```

```python
#get old and new values using new frame row id
for i in df_db_err.index:
    # replace values of one DataFrame with the value of another DataFrame
    df_db['position.lat'] = df_db['position.lat'].replace(df_db['position.lat'][i],np.nan)
    df_db['position.lon'] = df_db['position.lon'].replace(df_db['position.lon'][i],np.nan)
```

#### Save corrected file to folder

```python
# Create result csv file and save it
df_db.to_csv(result_file, sep=rslt_delimiter, encoding=rslt_encoding, index=False)
```

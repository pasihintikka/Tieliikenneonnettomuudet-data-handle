# Check all edited file column names

After edited files for inserting into SQL Server database, I notice that some newest files have more columns than older and the column order is different. So this code list columns and their orders in all files.

Code can use for one file type at a time (onnettomuus, alue, tieominaisuus, osallinen). Code create dataframe (and also file) contains list of filetype columns by years. So it can check easily which file columns needed to set same order with other files. End of the code is needed parts for that.

This check is important before inserting data into database tables without errors.

### Define parameters of the source and result files (needed for db table columns)

```python
# filename types are:
# tielonnett_2005_onnett.csv, tielonnett_2005_alue.csv, tielonnett_2005_tieomin.csv, tielonnett_2005_osall.csv
# File path and name
src_file_start = '../datasets/onnettomuus/db/tielonnett_'
src_file_end = '_onnett.csv'
years_start = 2005
years_end = 2021
result_file = '../datasets/onnettomuus/db/onnett_columns.csv'

# File attributes
src_delimiter = ';'
src_encoding = 'utf-8'
rslt_delimiter = ';'
rslt_encoding = 'utf-8'
```

```python
# Used libraries
import pandas as pd
import numpy as np
import math
```

#### Create empty dataframe with column named as years 

```python
names = []
for i in range(years_start,years_end+1):
    names.append(str(i))
            
df_res = pd.DataFrame(columns=names)
```

```python
list(df_res)
```

#### Check column amount of files and create needed row indexes to dataframe

```python
df_res = df_res.reset_index(drop=True, inplace=True)
idx = []
for i in range(years_start,years_end+1):
    df = pd.read_csv(src_file_start+str(i)+src_file_end, delimiter=src_delimiter, encoding=src_encoding)
    lst = list(df.columns)
    idx.append(len(lst))

df_res = pd.DataFrame(index = range(1, max(idx)+1, 1))
```

#### Save column list from the files to dataframe columns by years

```python
filler = 0
for i in range(years_start,years_end+1):
    df = pd.read_csv(src_file_start+str(i)+src_file_end, delimiter=src_delimiter, encoding=src_encoding)
    lst = list(df.columns)
    df_res[str(i)] = lst + [filler]*(len(df_res.index) - len(lst))
```

```python
df_res
```

#### Finishing

```python
# Create result csv file and save it
df_res.to_csv(result_file, sep=rslt_delimiter, encoding=rslt_encoding, index=False)
```
